```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('elysee.csv')
```

```python

df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-01')

# Aggregate events by date
calendar_df = df.groupby(['date']).agg(count=('desc', 'count')).reset_index()

# Extract year and month from date
calendar_df['year'] = calendar_df['date'].dt.year
calendar_df['month'] = calendar_df['date'].dt.month_name()

# Create the chart
chart = alt.Chart(calendar_df).mark_rect().encode(
    x=alt.X('date(date):O', title='Date'),
    y=alt.Y('month(date):O', title='Month'),
    color=alt.Color('count:Q', title='Count of Events', scale=alt.Scale(scheme='reds')),
    tooltip=[alt.Tooltip('date:T', title='Date'), alt.Tooltip('count:Q', title='Count of Events')]
).properties(
    width=700,
    height=300,
    title='Calendar Heatmap of Events'
)

chart




```
