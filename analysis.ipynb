```python


# assume that you have a DataFrame called "df" with a column "text"
# containing the text data you want to search for country names
# and a list "country_list" containing the country names in French

country_mentions = []

# iterate over each country name in the list
for country in pays:
    # check if the country name appears in any row of the "text" column
    country_mask = df['desc'].str.contains(country)
    # if the country name is mentioned in at least one row, add it to the list
    if country_mask.any():
        country_mentions.append(country)

# print the list of country names mentioned in the text data
print(country_mentions)

```

```python

# create an empty dictionary to store the count of mentions for each country
country_counts = {}

# iterate over each country name in the list
for country in pays:
    # check if the country name appears in any row of the "text" column
    country_mask = df['desc'].str.contains(country)
    # if the country name is mentioned in at least one row, add it to the dictionary
    if country_mask.any():
        country_counts[country] = country_mask.sum()

# print the dictionary of country mentions and their counts
print(country_counts)
```

```python
# order countries by number of mentions

# create a DataFrame from the dictionary

country_counts_df = pd.DataFrame.from_dict(country_counts, orient='index', columns=['count'])

# sort the DataFrame by the "count" column in descending order

country_counts_df.sort_values(by='count', ascending=False, inplace=True)
```
